```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('/content/train.csv')
train
```

```python
train.columns
```

```python
drop_columns = ['Case Number', 'Date', 'Block', 'IUCR', 'Description',
                'Domestic', 'Beat', 'Ward', 'Community Area', 'FBI Code',
                'X Coordinate', 'Y Coordinate', 'Latitude', 'Longitude', 'Location']
for i in drop_columns:
  train.drop(i, axis=1, inplace=True)
```

```python
#The amount of missing values is small so input the most frequency values in the string columns sound a good idea
string_column = ['Location Description', 'District', 'Primary Type', 'Arrest']
for col in string_column:
  most_frequency = train[col].mode()[0]
  train[col].fillna(most_frequency, inplace=True)

#Change the missing values on numeric column for the mean because tyhe same mition up
numeric_column = ['Month', 'Year', 'Day']
for col in numeric_column:
  mean = train[col].mean()
  train[col].fillna(mean, inplace=True)


#As we input mean values in the missing values place we get decimal values, month and year is date values so change from float to int
#help us have a vision more cleary and these decimal values can disrupt the model
train_numeric_column = ['Month', 'Year', 'Day']
for col in train_numeric_column:
  train[col] = train[col].astype('int')
```

```python
frequency = train['Primary Type']
frequency = pd.DataFrame(frequency)

frequency_mapping = frequency['Primary Type'].value_counts().to_dict()
train['Primary Type'] = train['Primary Type'].map(frequency_mapping)

local = train['Location Description']
local = pd.DataFrame(local)
local_mapping = local['Location Description'].value_counts().to_dict()
train['Location Description'] = train['Location Description'].map(local_mapping)
```

```python
negative_counts = (train < 0).sum()

print("Negative value counts in each column:")
print(negative_counts)
```

```python
column_name = 'District'
mean_value = train[train['District'] >= 0][column_name].mean()
train['District'] = np.where(train['District'] < 0, mean_value, train['District'])
```

```python
train_model = train.drop('Arrest', axis=1)
test_model = train['Arrest']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_model, test_model, train_size=0.7, random_state=42)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB

models = [RandomForestClassifier, DecisionTreeClassifier, MultinomialNB, GaussianNB, BernoulliNB]
for i in models:
  model = i()
  model.fit(x_train, y_train)
  y_pred = model.predict(x_test)
  accurancy = accuracy_score(y_test, y_pred)
  print(f"Score {i}: {accurancy}")
```

```python
from tensorflow import keras
from tensorflow.keras import layers

#Was created a sequencial model with 6 inputs layers one hidden leyer with 128 neurons and 1 output leyer
model = keras.Sequential([
    layers.Input(shape=(6,)),
    #The activation relu take us a linear values
    layers.Dense(128, activation='relu'),
    #The sigmoid activation sigmoid the the values back and compile the values between 0 and 1
    layers.Dense(1, activation='sigmoid')
])

#We compile the models and add parameters as optimizer, loos and metric
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
#We setting the epochs we going run this model and the size of batch
epochs = 10
batch_size = 64

#Now we test the model fit real values set the epochs set the bath and give the validatio datas
history = model.fit(x_train, y_train,
                    epochs=epochs,
                    batch_size=batch_size,
                    validation_data=(x_test, y_test))

#See the loss of the model
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f'test loss: {test_loss}, test_accurancy: {test_acc}')
```

```python
#we need transform these variables in arrays
x_train = x_train.values
x_test = x_test.values
y_train = y_train.values
y_test = y_test.values
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
#Using the torch to transform the value in numpy(float)
x_train = torch.from_numpy(x_train).float()
y_train = torch.from_numpy(y_train).float()
x_test = torch.from_numpy(x_test).float()
y_test = torch.from_numpy(y_test).float()

#create a dataset with tensor that this dataset join the train variables and labels train variable
train_dataset = torch.utils.data.TensorDataset(x_train, y_train)
#Load this dataset
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)

#create a dataset with tensor that this dataset join the test variables and labels train variable
test_dataset = torch.utils.data.TensorDataset(x_test, y_test)
#Load this dataset
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64)

#Our target is a booleam variable that is binary values so we defined a submodel on nn.module
class BinaryClassifier(nn.Module):
    #Create a def with the structure to the neural network
    def __init__(self, input_size):
        # create a variable to call the constructor parent class nn.module
        super(BinaryClassifier, self).__init__()
        # using linear leyers for this case
        self.fc1 = nn.Linear(input_size, 128)
        # Rectified Linear Unit(ReLu) is a linear activation neural that get us a output binary
        self.relu = nn.ReLU()
        #This leyer take the output before with 128 units and get us just one value
        self.fc2 = nn.Linear(128, 1)
        #This is the last layer that layer compress the last output in a value between 0 or 1
        self.sigmoid = nn.Sigmoid()

    #This def apply the datas in the parameters builded before
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x

#Put the model in a variable and set the numbers of input layers this case is 6 because we get 6 columns in the datset
model = BinaryClassifier(input_size=6)
#Using the fuction BCE Loss that return us a binary result
criterion = nn.BCELoss()
#setting a optmizer with Adam with a learning rate starting the 0.001 chaging the weight of variable to have the best score
optimizer = optim.Adam(model.parameters(), lr=0.001)


#Set the number of epochs the model will running
epochs = 10
for epoch in range(epochs):
    model.train()
    #Split the Dataloader in tow batches (x,y)
    for batch_x, batch_y in train_loader:
        #We clear each batch for don't have remnants of batch before
        optimizer.zero_grad()
        #Now the model calculate the outpust based in the batch_x input values
        outputs = model(batch_x)
        #calculate the loos between the precitic values and real value(batch_y)
        #the criterion it's a fuction loos and the view(-1) is these preditions have the same shape of real variable
        loss = criterion(outputs.view(-1), batch_y)
        #This line calcule the gradients of the model's parameters this allow the model change the paremeters
        loss.backward()
        #The all updates is apply here
        optimizer.step()


#Usinig the model for evaluation
model.eval()
#these 3 variables staring with 0 to see the performance of model the test loos storege the loos values
test_loss = 0.00
#The value of correct is the how much is the variance between the real values and predict values
correct = 0
#This variable have the shape of real target values batch_y
total = 0

#we disable the  temporarily the gradiente in the analyze
with torch.no_grad():
  #Split the test dataloder in two batch (x,y)
  for batch_x, batch_y in test_loader:
    #Predicting values based in the input values batch_x
    outputs = model(batch_x)
    #Calculating the loos between predict values and real values
    loss = criterion(outputs.view(-1), batch_y)
    #We put the loos values in this variable
    test_loss += loss.item()
    #The target is a boolean columns that is binary value(0 or 1) so we builded a filter where if the value is largest
    #then 0.5 this value is agreed 1
    predicted = (outputs >=0.5).squeeze().long()
    #This variable storege the shape of the real variable (batch_y)
    total += batch_y.size(0)
    #This variable have the loos values
    correct += (predicted == batch_y).sum().item()

print(f'Test loss: {test_loss / len(test_loader)}, Test accurancy: {100 * correct / total}%')
```
