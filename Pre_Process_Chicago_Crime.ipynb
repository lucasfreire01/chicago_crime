```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('/content/train.csv')
train
```

```python
train.columns
```

```python
drop_columns = ['Case Number', 'Date', 'Block', 'IUCR', 'Description',
                'Domestic', 'Beat', 'Ward', 'Community Area', 'FBI Code',
                'X Coordinate', 'Y Coordinate', 'Latitude', 'Longitude', 'Location']
for i in drop_columns:
  train.drop(i, axis=1, inplace=True)
```

```python
train
```

```python
#Take the district column and change for names of values to a better vision
district = {
    1.0 : 'Bedford',
    2.0 : 'South Side',
    3.0 : 'Pullman',
    4.0 : 'Cook County',
    5.0 : 'Roseland',
    6.0 : 'Auburn Gresham',
    7.0 : 'West Englewood',
    8.0 : 'Chicago Lawn',
    9.0 : 'Pilsen',
    10.0 : 'South Lawndale',
    11.0 : 'Near West Side',
    12.0 : 'Wicker Park',
    14.0 : 'Logan Square',
    15.0 : 'Austin',
    16.0 : 'North Mayfair',
    17.0 : 'Albany Park',
    18.0 : 'Goose Island',
    19.0 : 'Lakeview',
    20.0 : 'Lincoln Square',
    22.0 : 'Beverly',
    24.0 : 'Rogers Park',
    25.0 : 'Belmont Cragin'
}
train['District'] = train['District'].map(district)
```

```python
train
```

```python
#Transfor the Arrest column in numeric with One hot encoder
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
data_encode = train['Arrest'].values.reshape(-1,1)
encoder.fit(data_encode)
data_encode = encoder.transform(data_encode).toarray()
encoder_df = pd.DataFrame(data_encode, columns=encoder.get_feature_names_out(['Arrest']))

#Put the columns genereted in the original dataset
train['Arrest_False'] = encoder_df['Arrest_False']
train['Arrest_True'] = encoder_df['Arrest_True']

#Drop the Arrest column
train.drop('Arrest', axis=1, inplace=True)
```

```python
train
```

```python
train.isnull().sum()
```

```python
#The amount of missing values is small so input the most frequency values in the string columns sound a good idea
string_column = ['Location Description', 'District']
for col in string_column:
  most_frequency = train[col].mode()[0]
  train[col].fillna(most_frequency, inplace=True)

#Change the missing values on numeric column for the mean because tyhe same mition up
numeric_column = ['Month', 'Year']
for col in numeric_column:
  mean = train[col].mean()
  train[col].fillna(mean, inplace=True)


#As we input mean values in the missing values place we get decimal values, month and year is date values so change from float to int
#help us have a vision more cleary and these decimal values can disrupt the model
train_numeric_column = ['Month', 'Year', 'Arrest_False', 'Arrest_True']
for col in train_numeric_column:
  train[col] = train[col].astype('int')
```
