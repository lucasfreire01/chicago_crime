```python
import pandas as pd

# Sample data
data = dataset['IUCR']

# Create a DataFrame
df = pd.DataFrame(data, columns=['IUCR'])

# Create a new 'Category' column based on the condition
df['Category'] = df['IUCR'].apply(lambda x: 'ROBBERY' if x.startswith(('031', '032', '033')) else 'Other')

print(df)

```

```python
df
```

```python
filtered_df = df[df['IUCR'].str.startswith(('031', '032', '033'))]

print(filtered_df)
```

```python
import pandas as pd

# Sample data
data = dataset['IUCR'].unique()

# Create a DataFrame
df = pd.DataFrame(data, columns=['Crime_Code'])

# Extract the first digits and letters
df['Category'] = df['Crime_Code'].str.extract(r'(\d+)([A-Z]*)', expand=False).apply(tuple, axis=1)

# Group by Category and count occurrences
category_counts = df.groupby('Category').size().reset_index(name='Count')

print(category_counts)
```
