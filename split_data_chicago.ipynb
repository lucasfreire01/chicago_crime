```python
import pandas as pd
```

```python
dataset = pd.read_csv('/content/Chicago Total Crimes 2001 - 2023.csv')
```

```python
dataset.columns
```

```python
dataset
```

```python
#trated the date column split in day, month and year
dataset['Date'] = pd.to_datetime(dataset['Date'])
dataset['Day'] = dataset['Date'].dt.day
dataset['Month'] = dataset['Date'].dt.month
dataset['Year'] = dataset['Date'].dt.year
```

```python
dataset
```

```python
#The distribuation of year column
dataset['Year'].value_counts()
```

```python
#In the dataset the date values is messy so we organieze it based in the month and year
dataset = dataset.sort_values(by=['Year', 'Month'])
```

```python
dataset
```

```python
#As this dataset is so big we split in train and test, the train dataset have 60% of original dataset and test with 40% of original dataset
split_ratio = 0.6
split_idx = int(len(dataset) * split_ratio)

train = dataset[:split_idx]
test = dataset[split_idx:]

train.to_csv('train.csv', index=False)
test.to_csv('test.csv', index=False)
```
